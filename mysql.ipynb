```python
# 필요한 라이브러리 임포트
import numpy as np
import pandas as pd
from sqlalchemy import create_engine, text 

# sqlalchemy engine 생성 (DB 연결)
engine = create_engine("mysql+pymysql://root:eva1234@localhost:3306/library?charset=utf8mb4")
print(engine)

# 데이터 추출을 위한 SQL
search = 'c++'  # 책 제목에 대한 키워드 검색  
sql = "SELECT bisbn, btitle, bauthor, bprice FROM books WHERE btitle LIKE :keyword"

# with 사용하여 DB 연결 및 실행
with engine.connect() as connection:
    result = connection.execute(text(sql), {'keyword': f'%{search}%'})
    #print(result.fetchone())
    df=pd.DataFrame(result.fetchall())
display(df)
```
